```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA, KernelPCA
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df_business_service_2020 = pd.read_excel('business_service_2020.xlsx', index_col=0)
df_business_service_2021 = pd.read_excel('business_service_2021.xlsx', index_col=0)
```

```python
def minmax(y_weighted):
    return(y_weighted - y_weighted.min())/(y_weighted.max() - y_weighted.min()) * 50 + 50
```

```python
df_business_service_2020
```

```python
scaler = MinMaxScaler()
x = scaler.fit_transform(df_business_service_2020)
pca = PCA(.99)
y = pca.fit_transform(x)
print(y)
pca.explained_variance_ratio_.cumsum()
```

```python
y_weighted = np.dot(y, pca.explained_variance_ratio_,)
y_weighted
```

```python
x_row = df_business_service_2020.index.to_list()
#print(x_row)
score = minmax(y_weighted)
score
```

```python
scaler = MinMaxScaler()
x2021 = scaler.fit_transform(df_business_service_2021)
pca = PCA(.99)
y2021 = pca.fit_transform(x2021)
print(pca.singular_values_)
print(pca.components_)
print(y2021)
pca.explained_variance_ratio_.cumsum()
```

```python
y2021_weighted = np.dot(y2021, pca.explained_variance_ratio_,)
y2021_weighted
```

```python
score2021 = minmax(y2021_weighted)
score2021
```
