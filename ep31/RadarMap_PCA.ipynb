```python
for i in range(len(l)):
    score2020 = []
    score2021 = []
    for k in data2021.keys():
        score2020.append(data2020[k][i])
        score2021.append(data2021[k][i])
    score2020.append(score2020[0])
    score2021.append(score2021[0])
    theta = np.linspace(0, np.pi*2, len(dimensions))
    plt.figure(figsize=(12,9), facecolor='white')
    ax = plt.subplot(111, projection='polar')
    ax.set_title(l[i] + '市评分雷达图', fontsize=15)
    ax.set_rlim(40,100)
    ax.plot(theta, score2020, marker='o', linestyle='--', linewidth=1.5, alpha = 0.5, color='blue')
    ax.plot(theta, score2021, marker='o', linestyle='--', linewidth=1.5, alpha = 0.5, color='red')
    ax.set_thetagrids(theta*180/np.pi, dimensions)
    ax.fill(theta, score2020, alpha=0.2)
    ax.fill(theta, score2021, alpha=0.2)
    plt.legend(['2020','2021'])
    plt.savefig(f'cities_{i}_2020_2021.png', bbox_inches='tight')
    plt.show()
```

```python
#总分
df_2020 = pd.DataFrame(data2020, index=l)
df_2021 = pd.DataFrame(data2021, index=l)
res_mean_2020 = df_2020.mean(axis=1)
res_mean_2021 = df_2021.mean(axis=1)
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
```

```python
s = pd.DataFrame()
s['score'] = np.append(res_mean_2021, res_mean_2020)
s['分数'] = s['score']
s['city'] = l + l
s['城市'] = s['city']
s['year'] = [2021] * len(l) + [2020] * len(l)
s['年份'] = s['year']
```

```python
plt.figure(figsize=(15,12), facecolor='white')
bar = sns.barplot(x = '城市', y = '分数', hue = '年份', data = s, palette = 'Pastel1')
for i in range(len(bar.containers)):
    plt.bar_label(bar.containers[i], fontsize=12, fmt='%.2f', padding = 2)
plt.title('上海市及江苏省13个地级市总分排名',fontsize=16)

plt.savefig('cities_overall_rankings_2020_2021.png', bbox_inches='tight')
plt.show()
```
